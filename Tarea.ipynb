# **LISTA**

```python
lista = ["cangrejo", "coral", "almeja", "conchas", "arena"]
print(lista)
```

```python
tupla = tuple(lista)
print(tupla)
```

```python
conjunto = set(lista)
print(conjunto)
```

```python
lista = ["cangrejo", "coral", "almeja", "conchas", "arena"]
diccionario = dict(enumerate(lista))
print(diccionario)
```

# **TUPLA**

```python
mi_tupla = ("cangrejo", "coral", "almeja", "conchas", "arena")
print("Tupla :", mi_tupla)
```

```python
lista = list(mi_tupla)
print("Lista :", lista)
```

```python
conjunto = set(mi_tupla)
print("Conjunto :", conjunto)
```

```python
diccionario = dict(enumerate(mi_tupla))
print("Diccionario :", diccionario)
```

# **DICCIONARIO**

```python
diccionario = {
    "Caminar": "10 minutos",
    "Correr": "20 minutos",
    "Nadar": "20 minutos"
}
print(diccionario)
```

```python
Tupla = tuple(diccionario.items())
print(Tupla)
```

```python
lista = list(diccionario.items())
print(lista)
```

```python
Conjunto = set(diccionario.items())
print(Conjunto)
```
