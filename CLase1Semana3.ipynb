# **CONTROL DE FLUJO EN PYTHON**

```python
a = 0
b = 1
while b < 1000:
  print(b)
  a = b
  b = a+b
```

```python
a, b = 0, 1
while b < 100:
  print(b)
  a, b = b, a+b
```

nótese que en el primer código el elemnto b sí se actualiza, a diferencia del segundo que no se actualiza sino que toma el elemneto anterior(Fibonacci)

```python
a, b = 0, 1
for b in range(21):   # 6 repeticiones
    print(b)
    a, b = b, a + b
```
