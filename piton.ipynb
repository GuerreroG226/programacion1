# **CONTROL DE FLUJO EN PYTHON**

```python
a = 0
b = 1
while b < 1000:
  print(b)
  a = b
  b = a+b
```

```python
a, b = 0, 1
while b < 100:
  print(b)
  a, b = b, a+b
```

nótese que en el primer código el elemnto b sí se actualiza, a diferencia del segundo que no se actualiza sino que toma el elemneto anterior(Fibonacci)

```python
a, b = 0, 1
for b in range(21):   # 6 repeticiones
    print(b)
    a, b = b, a + b
```

SENTENCIA TIPO IF

```python
x = int(input("Ingresa un entero, por favor: "))
if x < 0:
  x = 0
  print('Negativo cambiado a cero')
elif x == 0:
  print('Cero')
elif x == 1:
    print('Uno')
else:
  print('Mayor que uno')
```

# La sentencia for en Python itera sobre los ítems de cualquier secuencia (una lista o una cadena de texto), en el orden que aparecen en la secuencia.

```python
for i in range(1, 6):
    print(i)
```

```python
import numpy as np
secuencia = np.linspace(0, 10, 50)
for i in secuencia:
    print(i)
```

```python
for z in range(-10, -100, -30):
  print(z)
```

```python
frutas = ["manzana", "banana", "pera"]
for fruta in frutas:
    print(fruta)
```

```python
colores = ("rojo", "verde", "azul")

for color in colores:
    print(color)
```

```python
a = ['Mary', 'tenia', 'un', 'corderito']
for i in range(len(a)):
  print(i, a[i])
```

```python
a = ['Mary', 'tenia', 'un', 'corderito']
for i, palabra in enumerate(a):
    print(i, palabra)
```

```python
a = ['Mary', 'tenia', 'un', 'corderito']
for i in enumerate(a):
    print(i)
```

```python
a = ['Mary', 'tenia', 'un', 'corderito']
for i, palabra in enumerate(a):
    print(i, ",", palabra)
```
